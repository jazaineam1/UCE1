# Crear archivos interactivos y cargar a la nube

Shiny es un paquete de R para la construcción de cuadros de mando web interactivos. Permite, por ejemplo, crear interfaces para algoritmos o acceder y manipular tablas de datos a través de controles de HMTL: sliders, botones, etc.\
<a href="url"><img src="https://diegokoz.github.io/intro_ds_bookdown/img/shiny_logo.png" align="center" height="100" width="100" ></a>

Primero debemos instalar el paquete `shiny` en nuestro Rstudio o Rstudio.cloud.

![](https://raw.githubusercontent.com/jazaineam1/UCE1/master/Cuadernos/Images/instalar_shiny.gif)




Preuba en R ejecutar las lineas 
 ```r 
library(shiny)
runExample(example = "01_hello")
```

![](https://github.com/jazaineam1/UCE1/blob/master/Cuadernos/Images/cargar_ejemplo.gif?raw=true)

Para más ejemplo visitar 
https://shiny.rstudio.com/gallery/

Cierra el cuadro creado, para terminar la ejecución de R.


Ahora instala flexdashboard
![](https://github.com/jazaineam1/UCE1/blob/master/Cuadernos/Images/flex_dash.gif?raw=true)



Crea un nuevo proyecto flexdashboard

![](https://github.com/jazaineam1/UCE1/blob/master/Cuadernos/Images/Abrir_flexdash.gif?raw=true)

Revise el enlace https://bookdown.org/yihui/rmarkdown/layout.html  y https://bookdown.org/yihui/rmarkdown/dashboard-components.html para  cambiar la configuración de presentación de su dashboard.


Ahora revise la página https://bookdown.org/yihui/rmarkdown/shiny.html#shiny emule el ejercicio de "Old Faithful Eruptions".

![](https://github.com/jazaineam1/UCE1/blob/master/Cuadernos/Images/shiny.gif?raw=true)





## Subir al servidor en la nube

Para esto se debe crear una cuenta en https://www.shinyapps.io/


Luego de que se creee la cuenta se deben instalar los paquetes para enlazar el servidor con Rstudio.

![](https://github.com/jazaineam1/UCE1/blob/master/Cuadernos/Images/shinyapp.gif?raw=true)

En nuestra página de https://www.shinyapps.io/ creamos el token de enlace que será pedido en Rstudio, lo pegamos y publicamos el app, esperamos unos minutos y hacer uso de la aplicación en cualquier momento.


![](https://github.com/jazaineam1/UCE1/blob/master/Cuadernos/Images/shinyapp_cargar.gif?raw=true)
